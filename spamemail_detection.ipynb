```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
dataset = pd.read_csv('/content/mail_data.csv')
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
#replacing null values with null string
mail_data = dataset.where(pd.notnull(dataset), '')
```

```python
mail_data.head()
```

```python
mail_data.shape
```

```python
#label encodeing 0 as spam and 1 as ham
mail_data.loc[mail_data['Category'] == 'spam', 'Category'] = 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category'] = 1
```

```python
#Separating category and message into x and y
X = mail_data['Message']
Y = mail_data['Category']
```

```python
print(X)
print(Y)
```

```python
#splitting test data and train data
Xtr,Xte,Ytr,Yte = train_test_split(X,Y, test_size = 0.2, random_state = 2)
```

```python
print(Xtr.shape,X.shape,Ytr.shape,Y.shape)
```

```python
# transform the text data to feature vectors that can be used as input to the Logistic regression

feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)

Xtr_features = feature_extraction.fit_transform(Xtr)
Xte_features = feature_extraction.transform(Xte)

# convert Y_train and Y_test values as integers

Ytr = Ytr.astype('int')
Yte = Yte.astype('int')
```

```python
print(Xtr_features)
```

```python
mode = LogisticRegression()
```

```python
mode.fit(Xtr_features,Ytr)
```

```python
#prediction if email is spam or ham on training data
predictOnTrainData = mode.predict(Xtr_features)
accuracyScore = accuracy_score(Ytr,predictOnTrainData)
```

```python
print("the accuracy score of this model is: ", accuracyScore*100)
```

```python
#prediction on test data
predictonTestData = mode.predict(Xte_features)
accuracyScoreOnTestData = accuracy_score(Yte,predictonTestData)
```

```python
print("the accuracy score of this model on test data is: ", accuracyScoreOnTestData*100)
```

**Building a predictive model**

```python
inputMail = ["There you have it—three great opportunities to learn and earn. Whether you're bringing your community onchain, reimagining cross-chain apps, or something beyond, we can’t wait to see what you’ll build.Register now: https://skate-stateless-innovation-program.devfolio.co/"]
#converting input mail to feature vector
inputMail_features = feature_extraction.transform(inputMail)

#making prediction
prediction = mode.predict(inputMail_features)


if(prediction[0]==1):
  print("this is ham mail")

else:
  print("this is a spam mail")
```
